```python
dt = pd.read_csv('forecast1.csv', sep=';')

result = []
while True:
    clear_output(wait=True)
    idx = randint(0, len(dt) - 1)
    dts = dt[idx - 3: idx + 1]
    dict_values = dts.T.to_dict()
    indexes = dts.index.to_list()
    print('Область: {}'.format(dt.loc[idx, 'areal']))
    print('Город: {}'.format(dt.loc[idx, 'city']))
    print('Месяц:      \t{}\t{}\t{}\t{}'.format(*[month_map[dict_values[index]['month']] for index in indexes]))
    print('День:      \t{}\t{}\t{}\t{}'.format(*[dict_values[index]['day'] for index in indexes]))
    print('Температура:\t{}\t{}\t{}\t{}'.format(*[dict_values[index]['day_temp'] for index in indexes]))
    print('Давление:\t{}\t{}\t{}\t{}'.format(*[dict_values[index]['day_pressure'] for index in indexes]))
    print('Облачность:\t{}\t{}\t{}\t{}'.format(*[obl_map[dict_values[index]['day_obl']] for index in indexes]))
    print('Явление:\t{}\t{}\t{}\t{}'.format(*[phen_map[dict_values[index]['day_phen']] for index in indexes]))
    print('Направление:\t{}\t{}\t{}\t{}'.format(*[dict_values[index]['day_dir'] for index in indexes]))
    print('Ветер:        \t{}\t{}\t{}\t{}'.format(*[dict_values[index]['day_wind'] for index in indexes]))
    for fish in fishs:
        print('{} ({}):'.format(fish, fish_rec[fish][dict_values[indexes[-1]]['month']]), end='')
        forecast = input()
        dt.loc[idx, fish] = forecast
        
```
