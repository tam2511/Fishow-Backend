```python
import requests
import matplotlib.pyplot as plt
from json import dumps
from tqdm.notebook import tqdm
from time import time
from random import choice
```

# Получение информации о регионах

```python
st = time()
r = requests.get('https://back.fishow.ru/api/region/')
print('request time: {:.3f} sec'.format(time() - st))
print('status code: {}'.format(r.status_code))
print('json response:')
print('======================================')
print(dumps(r.json(), indent=4))
```

# Поиск регионов

```python
search_set = list("абвгдежзийклмнопстуфхцчшщъыьэюя")
len_ = len(search_set)
# 2
for i in range(len_):
    for j in range(i, len_):
        search_set += [search_set[i] + search_set[j]]
        for k in range(j, len_):
            search_set += [search_set[i] + search_set[j] + search_set[k]]
```

## Пример поиска

```python
st = time()
search_param = 'м'
r = requests.get('https://back.fishow.ru/api/region/?search={}'.format(search_param))
print('request time: {:.3f} sec'.format(time() - st))
print('status code: {}'.format(r.status_code))
print('json response:')
print('======================================')
print(dumps(r.json(), indent=4))
```

## Нагрузка на поиск

```python
times = {1: [], 2: [], 3: []}
for search_param in tqdm(search_set):
    len_ = len(search_param)
    st = time()
    r = requests.get('https://back.fishow.ru/api/region/?search={}'.format(search_param))
    times[len_].append(time() - st)
    
plt.plot(list(range(len(times[1]))), times[1], label="Поиск одного символа")
plt.plot(list(range(len(times[2]))), times[2], label="Поиск двух символа")
plt.plot(list(range(len(times[3]))), times[3], label="Поиск трех символа")
plt.title("Распределене скорости поиска")
plt.ylabel("Время запроса")
plt.legend(loc="upper left")
plt.show()
```
