```python
import requests
import matplotlib.pyplot as plt
from json import dumps
from tqdm.notebook import tqdm
from time import time
from random import choice
```

# Фильтрация прогноза для 10 суток по региону

```python
areals = ['', 'Московская область', 'Владимирская область', 'Липецкая область', 'Архангельская область', 'Республика Алтай']
url_pattern = 'https://back.fishow.ru/api/predictionten/?areal={}&date={}&city={}&fish={}'
times = []
for areal in areals:
    st = time()
    r = requests.get(url_pattern.format(areal, '', '', ''))
    times.append(time() - st)
    
print('times: {}'.format(times))
print('status code: {}'.format(r.status_code))
print('json response:')
print('======================================')
print(dumps(r.json(), indent=4))
```
